```python
selected_features = ['genres', 'id', 'keywords', 'title', 'overview', 'cast', 'director', 'original_language']
# I am keeping id for making the frontend
df = df[selected_features]
df.head()
```

```python
print(df.shape)
print('########')
print(df.info())
print('########')
print(df.isnull().sum())
```

```python
df.fillna('', inplace=True)
print(df.isnull().sum())
print('*********************')
print(df.info())
```

```python
df[['cast']].head()
```

```python
def cast_selector(cast):
    cast_list = cast.split(' ')
    misc = []
    for i in range(0,10):
        misc.append(str(i))
    y = []
    for i in cast_list:
        if cast_list.index(i)%2 == 0:
            try:
                name = i + cast_list[cast_list.index(i)+1]
                for j in misc:
                    if j in i or j in cast_list[cast_list.index(i)+1]:
                                                break
                    else:
                        y.append(name)
                        break
                
            except:
                break
            
    return y[0:3]
# used to filter and clean cast column
```

```python
pd.set_option('max_colwidth', 100)
df.cast = df['cast'].apply(cast_selector)
```

```python
df.cast = df['cast'].apply(lambda x: " ".join(x))
```

```python
df['tags'] = df['genres']+' '+df['keywords']+' '+df['overview']+' '+df['cast']+' '+df['director']+' '+df['original_language']
```

```python
ndf = df.drop(columns=['genres', 'keywords', 'overview', 'cast', 'director', 'original_language'])
ndf.head()
```

```python
ps = PorterStemmer()
def stem(tag):
    new_tag = []
    for i in tag.split(' '):
        new_tag.append(ps.stem(i))
    return ' '.join(new_tag)
ndf.tags = ndf['tags'].apply(stem)
# stemming all the tags
```

```python
ndf.head()
```

```python
cv = CountVectorizer(stop_words='english', max_features=7000)
vtags = cv.fit_transform(df['tags']).toarray()
```

```python
vtags.shape
```

```python
ndf.shape
```

```python
from sklearn.metrics.pairwise import cosine_similarity
# to calculate cosine distance and compare
```

```python
similarity = cosine_similarity(vtags)
similarity
```

```python
def recommend(movie):
    index = ndf[ndf['title'] == movie].index[0]
    recommedations = sorted(enumerate(similarity[index]), reverse = True, key=lambda x:x[1])[1:11]
    l = []
    j = 1
    print('Movies recommended for you are: ')
    for i in recommedations:
        print(f"{j}. {ndf.iloc[i[0], 1]}")
        j += 1
recommend('The Lego Movie')
```

```python
# you can recommend using this recommend function and passing the name of movie you like
# for example
# I am recommending 10 movies
recommend('Superman')
```
